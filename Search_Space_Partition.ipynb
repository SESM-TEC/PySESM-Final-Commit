```python
from re import sub
import itertools
import numpy as np
import torch
```

```python
def generate_partitions(a, b, T, n):

    partitions = [np.linspace(a, b, T+1) for _ in range(n)]

    corners = np.array(np.meshgrid(*partitions)).T.reshape(-1, n)

    return corners
```

```python
def assign_to_subblock(x, corners):

    n_dimensions = x.shape[0]

    sub_intervals = []

    sorted_corners = np.sort(corners, axis=0)

    for i in range(n_dimensions):

        index = np.digitize(x[i], sorted_corners[:, i]) - 1

        sub_interval = (sorted_corners[index, i], sorted_corners[index + 1, i])

        sub_intervals.append(sub_interval)

    return sub_intervals
```

```python
a = 0
b = 1
T = 4
n = 3

corners = generate_partitions(a, b, T, n)

x_observed = np.array([0.3, 0.6, 0.4])

assigned_intervals = assign_to_subblock(x_observed, corners)

for i, interval in enumerate(assigned_intervals):
    print(f"t_{i+1}: {interval}")
```
